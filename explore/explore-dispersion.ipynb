```python
import sys
sys.path.append('..')
import numpy as np
from scipy import constants as C
from matplotlib import pyplot as plt
import modules.dispersion as dispersion
```

```python
Npts = 300
# n00 = 0.0
# mult = 1.0
n00 = 1.000273
mult = 1e7
mks_length = 1.0e-6 / (2*np.pi)
um = 1e6*mks_length
cm = 100*mks_length
#mat = dispersion.ZnSe(mks_length)
mat = dispersion.HumidAir(mks_length,0.4,rel_linewidth=1e-4)
# mat.add_opacity_region(1/cm,.01/um,0.25/um)
def getw(lambda_mks):
    return 2*np.pi*mks_length/lambda_mks
def make_plots(wavelength_range):
    w = getw(wavelength_range)
    chi = mat.chi(w)
    nrefr = np.sqrt(1+chi)
    alpha = np.imag(nrefr)*w/mks_length
    plt.figure(1)
    plt.plot(wavelength_range*1e6,mult*(np.real(nrefr)-n00))
    plt.xlabel('Wavelength (um)')
    plt.ylabel(r'$(n-n_0)\times C$')
    plt.figure(2)
    plt.plot(wavelength_range*1e6,alpha)
    plt.xlabel('Wavelength (um)')
    plt.ylabel(r'$\alpha$ (m$^{-1}$)')
    plt.show()

#print(air.WaterVaporDensity())
```

```python
make_plots(np.linspace(0.15e-6,0.7e-6,Npts))
```
