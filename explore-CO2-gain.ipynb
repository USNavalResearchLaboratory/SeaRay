```python
import numpy as np
from matplotlib import pylab as plt
from scipy import constants as C
import co2amp as co2

isotopologues = [1.0] + [0.0]*5
N2 = 1.0
He = 6.0
mix = np.array(isotopologues + [N2,He])
T0 = 300.0
Temp2 = 300.0
Temp3 = 1000.0
```

```python
Nw = 1024
w = np.linspace(2*np.pi*C.c/8e-6,2*np.pi*C.c/12e-6,Nw)
icm = w/C.c/2/np.pi/100
G = co2.gain(w,mix,T0,Temp2,Temp3)
plt.figure(figsize=(10,4))
plt.plot(icm,np.real(G))
plt.plot(icm,np.imag(G))
plt.xlabel('Wavenumbers (cm$^{-1}$)')
plt.ylabel('Gain (m$^{-1}$)')
```
